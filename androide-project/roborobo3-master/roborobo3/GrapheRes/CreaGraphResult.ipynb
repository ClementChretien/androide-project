```python
import matplotlib.pyplot as plt
import operator
import copy
import numpy as np
```

```python
def getFileTab(f):
    fp = open(f,'r')
    line = fp.readline()
    d = dict()
    i = 0
    while line:
        if not('R' in line):
            comb,val = line.split(" ") 
            d[comb]=int(val[:-1])
        line = fp.readline()
    fp.close()
    return d
```

```python
dic = getFileTab("ResultatEval8.txt")
dic
```

```python
def creaHistArray(dic):
    
    array = []
    ar=[]
    ab=[]
    ac=[]
    ah=[]
    score = []
    while (len(dic.keys())>0 and len(score) < 100):
        maximum = max(dic, key=dic.get) 
        r,b,c,h = maximum.split("/")
        ar.append(int(r))
        ab.append(int(b))
        ac.append(int(c))
        ah.append(int(h))
        array.append(maximum)
        score.append(dic[maximum])
        del dic[maximum]
    print(array)
    return array,ar,ab,ac,ah,score
```

```python
def histogram(dic):
    ar,r,b,c,h,s = creaHistArray(copy.deepcopy(dic))
    header = ['Random','Bas','Complet','Haut']
    
    X_AXIS = s

    plt.rc('font', serif='Helvetica Neue')
    plt.rc('text', usetex='false')
    plt.rcParams.update({'font.size': 40})

    fig = plt.gcf()
    fig.set_size_inches(18.5, 5.5)

    configs = r
    N = len(r)
    ind = np.arange(N)
    width = 0.8
    
    bars = np.add(r, b).tolist()
    bars = np.add(bars,c).tolist()

    p1 = plt.bar(ind, r, width, color='r')
    p2 = plt.bar(ind, b, width, bottom=r, color='b')
    p3 = plt.bar(ind, c, width, bottom=b, color='g')
    p4 = plt.bar(ind, h, width, bottom=bars, color='y')

    #plt.ylim([0,120])
    plt.yticks(fontsize=12)
    plt.ylabel("Arrangement", fontsize=20)
    plt.xticks(ind, X_AXIS, fontsize=12, rotation=90)
    plt.xlabel('test', fontsize=12)
    plt.legend((p1[0], p2[0], p3[0], p4[0]), (header[0], header[1], header[2], header[3]), fontsize=12, ncol=4, framealpha=0, fancybox=True)
    plt.show()
```

```python
dic = getFileTab("ResultatEval14.txt")
histogram(dic)
```
