```python
import matplotlib.pyplot as plt
import operator
import copy
import numpy as np
```

```python
def getFileTab(f):
    fp = open(f,'r')
    line = fp.readline()
    r=0
    b=0
    c=0
    h=0
    while line:
        if ('R' in line):
            comb,val = line.split(" ") 
            r=max(float(val[:-1]),0)
        if ('B' in line):
            comb,val = line.split(" ") 
            b=float(val[:-1])
        if ('C' in line):
            comb,val = line.split(" ") 
            c=float(val[:-1])
        if ('H' in line):
            comb,val = line.split(" ") 
            h=float(val[:-1])
        line = fp.readline()
    fp.close()
    return r,b,c,h
```

```python
def getAllContrib(i,j):
    
    res = dict()
    for k in range(i,j+1):
        res[k] = getFileTab("CalculContribMarginale_"+str(k)+".txt")
    return res
```

```python
def creaHistArray(dic):
    print(dic)
    keys = []
    ar=[]
    ab=[]
    ac=[]
    ah=[]
    for i in range(1,len(dic)+1):
        r,b,c,h = dic[i]
        ar.append(round(r,2))
        ab.append(round(b,2))
        ac.append(round(c,2))
        ah.append(round(h,2))
        keys.append(i)
        
    return keys,ar,ab,ac,ah
```

```python
def soloHist(i,r,b,c,h):
    x=[r,b,c,h]
    #plt.hist(x, normed=True, bins=30)
    plt.ylabel('Contribution \n Marginale',size=25);
    plt.xlabel('Nombre agent : '+str(i),size=25);
    plt.bar([1,2,3,4], height= x,color=['red', 'green', 'blue', 'yellow'])
    plt.xticks([1,2,3,4], ['r','b','c','h'])
    plt.show()
```

```python
soloHist(1,1,8,3,4)
```

```python
def allGraph(i,j,nmax=100):
    dic = getAllContrib(i,j)
    for k in range(i,j+1):
        r,b,c,h = dic[k]
        soloHist(k,r,b,c,h)
```

```python
allGraph(1,9)
```
